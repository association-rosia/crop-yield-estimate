```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('../data/raw/digital-green-crop-yield-estimate-challenge20230912-4562-1bbmstk/Train.csv', index_col=0)
df
```

# Target: `Yield`

Yield of the current crop

```python
px.histogram(df, x='Yield', title='Yield')
```

```python
yield_by_acre = df['Yield'] / df['Acre'] 
px.histogram(x=yield_by_acre, title='Yield / Acre')
```

# Acre

Area of land under cultivation in acres

```python
Acre = df['Acre']

Acre.isna().value_counts()
```

```python
px.histogram(x=Acre, title='Histogram of Acre')
```

```python
fig = px.scatter(x=Acre, y=df['Yield'], title='Scatter plot of yield/acre by Acre')
fig.update_xaxes(title_text="Acre")
fig.update_yaxes(title_text="Yield per Acre")
fig.show()
```

# Harvest date

Date of harvesting

```python
Harv_date = pd.to_datetime(df['Harv_date'])
print('Number of Nan:')
Harv_date.isna().value_counts()
```

```python
px.histogram(x=Harv_date)
```

```python
display(Harv_date.dt.month.value_counts())

px.box(x=Harv_date.dt.month, y=yield_by_acre, title='Boxplot of yield/acre by month')
```

```python
Harv_date.dt.year.value_counts()
```
